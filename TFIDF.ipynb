```python
import pandas as pd
import numpy as np
import jieba
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = law.data.read_law(n=100,preprocessing=True).return_data()
```

```python
data['defendant']
```

```python
dat = data[data['process']!=''][['process','plantiff','defendant']]
```

```python
cutted = []
dic = dict()

for j in dat.index:
    cut = [i for i in jieba.cut(dat.loc[j]['process'])]
    for word in cut:
        try:
            dic[word] += 1
        except:
            dic[word] = 1
    cutted.append(cut)
```

```python
dictionary = pd.DataFrame.from_dict(dic, orient = 'index')
dictionary.head()
```
