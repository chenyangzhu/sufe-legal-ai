```python
import law.data
import pandas as pd
```

```python
import law.utils
```

```python
import jieba
```

```python
data = law.data.read_law(n=10000).return_data()
```

```python
data = data[['procedure','process','opinion','result','notice','plantiff','defendant','third_party']]
```

```python
word_bank = dict()
cutted = []
for j in range(len(data)):
    print(j/len(data)*100,"%")
    new_cut = []
    new_cut.extend([i for i in jieba.cut(data.iloc[j]['procedure'])])
    new_cut.extend([i for i in jieba.cut(data.iloc[j]['process'])])
    new_cut.extend([i for i in jieba.cut(data.iloc[j]['opinion'])])
    new_cut.extend([i for i in jieba.cut(data.iloc[j]['result'])])
    new_cut.extend([i for i in jieba.cut(data.iloc[j]['notice'])])
    for each in new_cut:
        if each != data.iloc[j]['plantiff'] and each != data.iloc[j]['defendant'] and each != data.iloc[j]['third_party']:
            try:
                word_bank[each] += 1
            except KeyError:
                word_bank[each] = 1
```

```python
dictionary = pd.DataFrame.from_dict(word_bank,orient="index")
```

```python
dictionary = dictionary.sort_values(by=0,ascending=False)
```

```python
dictionary.reset_index(inplace=True) # change index to a column
```

```python
def string2float(string):
    '''
    If the string is convertable to int, return the int.
    If not, return the string back.
    '''
    try:
        return float(string)
    except ValueError:
        return string
```

```python
# Delete all int types
dictionary = dictionary.applymap(string2float) # change int-type string to real int
dictionary = dictionary[dictionary.applymap(type)['index'] != float]
```

```python
# Delete all signs
sign_list = ['，',
             '（',
             '）',
             '、',
             '。',
             '“',
             '”',
             '；',
             '》',
             '《',
             '？',
             '！',
             ':',
             '…',
             ',',
             '.',
             '!',
             '?',
             '×',]

dictionary = dictionary[-dictionary.isin(sign_list)['index']]
```

```python
for i in dictionary.index:
    print(i/len(dictionary))
    if "市" in dictionary.loc[i]['index']:
        dictionary = dictionary.drop(i)
    elif "区" in dictionary.loc[i]['index']:
        dictionary = dictionary.drop(i)
    elif "公司" in dictionary.loc[i]['index']:
        dictionary = dictionary.drop(i)
    elif "集团" in dictionary.loc[i]['index']:
        dictionary = dictionary.drop(i)
```

```python
dictionary['frequency'] = dictionary[0] / sum(dictionary[0])
```
